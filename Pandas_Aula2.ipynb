```python
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df5.head()
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
#Verificando o tipo de dado de cada caluna
df.dtypes
```

```python
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Consultando linhas com valoers faltantes
df.isnull().sum()
```

```python
 #Substituindo os valores nulos pela m√©dia
 df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apaganda as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos de apenas 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
#Criando nova coluna "receita"
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
 df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
  #nlargest
  df.nlargest(3, "Receita")
```

```python
#nsmallest
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

# **Trabalhando com datas**

```python
#Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando a coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o m√™s e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
 # Retornando a data mais antiga
 df["Data"].min()
```

```python
#Calculando a difenren√ßa de dias
df["diferen√ßa_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando coluna do trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do m√™s de mar√ßo
vendas_mar√ßo_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
